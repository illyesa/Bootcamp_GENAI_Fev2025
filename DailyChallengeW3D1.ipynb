```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0) # j'ai utilisé une seed pour avoir a chaque fois la même chose et pouvoir comprend se qui se modifie quand je fait une modification
temperature = np.random.uniform(-5, 35, (10, 12))

cities = [f"City {i+1}" for i in range(10)]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
df = pd.DataFrame(temperature, index=cities, columns=months)

avg_temp = df.mean(axis=1)
highest_avg_temp_city = avg_temp.idxmax()
lowest_avg_temp_city = avg_temp.idxmin()

print(f"Average:\n{avg_temp}, \nHighest: {highest_avg_temp_city}, Lowest: {lowest_avg_temp_city}")
```

```python
plt.figure(figsize=(12, 6))
for city in cities:
    plt.plot(months, df.loc[city], label=city)

plt.title("Monthly Temperature for Each City")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
