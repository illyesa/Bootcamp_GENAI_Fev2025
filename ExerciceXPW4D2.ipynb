```python
# 2
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()

input_nodes = ['x1', 'x2', 'x3']
hidden_nodes = ['h1', 'h2', 'h3', 'h4']
output_nodes = ['o1', 'o2']

for i in input_nodes:
    for h in hidden_nodes:
        G.add_edge(i, h)

for h in hidden_nodes:
    for o in output_nodes:
        G.add_edge(h, o)

pos = {
    'x1': (0, 0), 'x2': (0, 1), 'x3': (0, 2),
    'h1': (1, 0), 'h2': (1, 1), 'h3': (1, 2), 'h4': (1, 3),
    'o1': (2, 0), 'o2': (2, 1)
}
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_weight="bold", arrows=True)
plt.title("Artificial Neural Network (ANN) Example")
plt.show()
```

```python
# 3
import numpy as np
import matplotlib.pyplot as plt

# Fixer la seed pour la reproductibilité
np.random.seed(0)

x = np.arange(-1, 1, 0.1)
y = -x**2 + np.random.normal(0, 0.05, len(x))

plt.scatter(x, y, color='blue', label='Data with noise')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Generated Dataset with Noise')
plt.legend()
plt.show()
```

```python
x_train = x[:12]
y_train = y[:12]
x_test = x[12:]
y_test = y[12:]

print(f"Training set: {list(zip(x_train, y_train))}")
print(f"Test set: {list(zip(x_test, y_test))}")
```

```python
# 4
import numpy as np
import matplotlib.pyplot as plt

def polynomial_fit(degree, x_train, y_train):
    coeffs = np.polyfit(x_train, y_train, degree)
    poly = np.poly1d(coeffs)
    return poly

def plot_polyfit(degree, x_train, y_train, x_test, y_test):
    poly = polynomial_fit(degree, x_train, y_train)
    plt.scatter(x_train, y_train, color='blue', label='Training data')
    plt.scatter(x_test, y_test, color='green', label='Test data')
    plt.plot(np.linspace(-1, 1, 100), poly(np.linspace(-1, 1, 100)), color='red', label=f'Poly degree {degree}')
    plt.legend()
    plt.title(f'Polynomial Fit (Degree {degree})')
    plt.show()

for degree in [1, 7, 11]:
    plot_polyfit(degree, x_train, y_train, x_test, y_test)
```

```python
# 5
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

train_rmse = []
test_rmse = []

for degree in range(1, 12):
    poly = polynomial_fit(degree, x_train, y_train)
    y_train_pred = poly(x_train)
    y_test_pred = poly(x_test)
    train_rmse.append(np.sqrt(mean_squared_error(y_train, y_train_pred)))
    test_rmse.append(np.sqrt(mean_squared_error(y_test, y_test_pred)))

plt.plot(range(1, 12), train_rmse, label='Train RMSE')
plt.plot(range(1, 12), test_rmse, label='Test RMSE')
plt.yscale('log')
plt.xlabel('Polynomial Degree')
plt.ylabel('RMSE (Log Scale)')
plt.legend()
plt.title('RMSE for Polynomial Degrees')
plt.show()
```
