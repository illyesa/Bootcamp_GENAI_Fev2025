```python
import pandas as pd
import gzip

# Charger IMDb title.basics (informations générales)
df_basics = pd.read_csv("title.basics.tsv.gz", sep='\t', compression='gzip', low_memory=False)

# Charger IMDb title.ratings (notes)
df_ratings = pd.read_csv("title.ratings.tsv.gz", sep='\t', compression='gzip', low_memory=False)

print(df_basics.head(), df_ratings.head())
```

```python
merged_df = pd.merge(df_basics, df_ratings, on='tconst')
print(merged_df.head())
```

```python
merged_df2 = pd.merge(merged_df, df_imdb, left_on='primaryTitle', right_on='Series_Title')
print(merged_df2.head())
```

```python
# Suppression des doublons
merged_df2.drop_duplicates(inplace=True)
# Vérification des valeurs manquantes
print(merged_df2.isnull().sum())
# Gestion des valeurs manquantes (par exemple, suppression des lignes avec des valeurs manquantes critiques)
merged_df2.dropna(subset=['Certificate', 'Meta_score','Gross'], inplace=True)
# Vérification du type des colonnes
print(merged_df2.dtypes)
```

```python
merged_df2.info()
```

```python
# Ajout de la colonne platform pour savoir sur quel platform on peut voir les film/série
# Ajoute de manière aléatoir car on a pas trouver de database ou d'information suffisament détaillé pour notre exemple
import  numpy as np
# Création du dictionnaire des proportions
platforms_dict = {
    'Netflix': {'films': 5000, 'series': 2600},
    'Hulu': {'films': 1200, 'series': 1300},
    'Prime Video': {'films': 20000, 'series': 2500},
    'Disney+': {'films': 1500, 'series': 700}
}

# Extraction des plateformes et calcul des proportions globales
platforms = list(platforms_dict.keys())
total_content = sum(sum(v.values()) for v in platforms_dict.values())  # Somme totale films + séries
probabilities = [sum(v.values()) / total_content for v in platforms_dict.values()]  # Proportions normalisées

# Attribution des plateformes dans merged_df2 selon les proportions
merged_df2["Platform"] = np.random.choice(platforms, size=len(merged_df2), p=probabilities)
merged_df2.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrage des films et séries avec une note IMDB ou averageRating supérieures à 8
high_rated = merged_df2[(merged_df2['averageRating'] >= 8.0) | (merged_df2['IMDB_Rating'] >= 8.0)]
# Comptage des genres parmi les films/séries les mieux notés
genre_counts = high_rated['genres'].value_counts().head(10)
# Création du graphique
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, hue=genre_counts.index, palette='viridis', legend=False)
plt.title('Top 10 des genres les mieux notés (averageRating ou IMDB_Rating > 8)')
plt.xlabel('Genre')
plt.ylabel('Nombre de films/séries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Combinaison des années 'Released_Year' et 'startYear'
merged_df2['Year_combined'] = merged_df2['Released_Year'].fillna(merged_df2['startYear'])
# Combinaison des notes 'averageRating' et 'IMDB_Rating'
merged_df2['average_combined'] = merged_df2['averageRating'].fillna(merged_df2['IMDB_Rating'])
# Calcul de la moyenne par année combinée
average_ratings_by_year = merged_df2.groupby('Year_combined')['average_combined'].mean()
# Création du graphique
plt.figure(figsize=(12, 6))
# Tracer la ligne pour 'average_combined'
sns.lineplot(x=average_ratings_by_year.index, y=average_ratings_by_year.values, marker='o', label='Note moyenne combinée', color='blue')
# Rotation des années sur l'axe X pour les rendre plus lisibles
plt.xticks(rotation=90)
# Titres et labels
plt.title('Note moyenne des films/séries par année de sortie (average_combined)')
plt.xlabel('Année de sortie')
plt.ylabel('Note moyenne')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
# Comptage du nombre de films et séries par plateforme
platform_counts = merged_df2.groupby(['Platform', 'Genre']).size().unstack(fill_value=0)
# Préparation des données
platforms = platform_counts.index
films = platform_counts.get('Film', default=0)  # Récupère les films ou 0 si absent
series = platform_counts.get('Série', default=0)  # Récupère les séries ou 0 si absent
# Création du graphique
plt.figure(figsize=(14, 6))
bar_width = 0.4  # Largeur des barres
# Positions des barres (deux groupes séparés)
n_platforms = len(platforms)
group_spacing = 1  # Espace entre les deux groupes
pos_films = list(range(n_platforms))  # Positions pour les films
pos_series = [x + n_platforms + group_spacing for x in range(n_platforms)]  # Positions pour les séries
# Barres pour les films (bloc 1)
plt.bar(pos_films,
        films,
        bar_width,
        color='skyblue',
        label='Films')
# Barres pour les séries (bloc 2)
plt.bar(pos_series,
        series,
        bar_width,
        color='orange',
        label='Séries')
# Ajout des titres et labels
plt.title('Nombre de films et séries disponibles par plateforme de streaming')
plt.xlabel('Plateforme')
plt.ylabel('Nombre')
plt.xticks(pos_films + pos_series,
          list(platforms) + list(platforms),  # Double les noms des plateformes
          rotation=45,
          ha='right')
# Ajout de la légende
plt.legend(title='Type')
# Ajustement de l'affichage
plt.tight_layout()
plt.show()
```
