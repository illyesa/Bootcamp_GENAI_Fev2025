```python
import pandas as pd

df = pd.read_csv("datascience_salaries.csv")
df.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['normalized_salary'] = scaler.fit_transform(df[['salary']])


df_encoded = pd.get_dummies(df, columns=['job_title', 'job_type', 'experience_level', 'location'], drop_first=True)

df_features = df_encoded.drop(columns=['Unnamed: 0', 'salary_currency', 'salary'])

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_features)
pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])

final_df = pd.concat([df, pca_df], axis=1)

salary_stats = df.groupby('experience_level')['salary'].agg(['mean', 'median']).reset_index()
salary_stats.columns = ['experience_level', 'average_salary', 'median_salary']

print(salary_stats)
```
