```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, color='green', linestyle='dashed', linewidth=2, marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```

```python
categories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [20, 35, 30, 15]

plt.bar(categories, values)
plt.title('Fruit Sales Data')
plt.xlabel('Fruit')
plt.ylabel('Sales')
plt.show()
```

```python
plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'], edgecolor='black', hatch='/')
plt.title('Customized Fruit Sales Data')
plt.xlabel('Fruit')
plt.ylabel('Sales')
plt.show()
```

```python
import numpy as np

data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
brands = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others']
market_share = [40, 30, 15, 10, 5]

plt.figure(facecolor="white")
plt.pie(market_share, labels=brands, autopct='%1.1f%%')
plt.title('Smartphone Market Share 2023')
plt.show()
```

```python
explode = (0.1, 0, 0, 0, 0)
plt.figure(facecolor="white")
plt.pie(market_share, labels=brands, autopct='%1.1f%%', explode=explode)
plt.show()
```

```python
import numpy as np

data = np.random.randn(100)
plt.boxplot(data)
plt.title('Basic Box Plot')
plt.ylabel('Values')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot([1, 2, 3], [1, 4, 9])
ax1.set_title('First Subplot')
ax2.plot([1, 2, 3], [1, 2, 3])
ax2.set_title('Second Subplot')
plt.show()
```

```python
categories=["PA", "PB", "PC", "PD", "PE"]
values=[10,45,5,14,27]
plt.bar(categories, values, color=['red', 'green', 'blue', 'purple'], edgecolor='black', hatch='/')
plt.title('Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
data = np.random.normal(0, 1, 500)
plt.hist(data, bins=10, edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
brands = ['work', 'sleep', 'leisure', 'exercise', 'Others']
market_share = [40, 30, 15, 10, 5]
mycolors = ["black", "hotpink", "b", "#4CAF50", "green"]

plt.figure(facecolor="white")
# plt.pie(market_share, labels=brands, autopct='%1.1f%%', shadow = True)
plt.pie(market_share, labels = brands, colors = mycolors, shadow = True)
plt.title('time spent on different daily activities')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data['total_bill'], bins=20, color='skyblue')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], fill=True, color='purple')
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter', color='green')
plt.show()
```
