```python
iris = pd.read_csv("Iris.csv")
print(iris)
```

```python
quali_c = iris.select_dtypes(include=["object"]).columns
quanti_c = iris.select_dtypes(include=["int64", "float64"]).columns
print(quali_c)
print(quanti_c)
```

```python


mean_value = iris["SepalLengthCm"].mean()
median_value = iris["SepalLengthCm"].median()
mode_value = iris["SepalLengthCm"].mode()[0]

print(f"Moyenne : {mean_value}")
print(f"Médiane : {median_value}")
print(f"Mode : {mode_value}")

# la moyenne de SepalLengthCm est de 5,843 sa median est 5,8 et le plus utiliser est 5
#trend analysis : Year, Avg hrs per day sleeping            pour voir si les Américains dorment plus ou moins qu'avant
#group comparison : Age Group, Avg hrs per day sleeping     pour voir si les jeunes dorment plus que les adultes
#                   Sex, Avg hrs per day sleeping           pour vérifier si les femmes dorment plus que les hommes
#                   Type of Days, Avg hrs per day sleeping  pour voir si les gens dorment plus le week-end
```
